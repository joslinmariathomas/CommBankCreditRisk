```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
application_train = pd.read_csv('../data/application_train.csv')
application_train.head()
```

```python
previous_application = pd.read_csv('../data/previous_application.csv')
previous_application.head()
```

```python
bureau = pd.read_csv('../data/bureau.csv')
bureau.head()
```

```python
application_null = (
    application_train.isnull().sum()
        .to_frame(name='sum_of_nulls')
        .assign(percent_null=lambda d: d['sum_of_nulls'] * 100 / len(application_train))
        .reset_index(names='column')
        .sort_values('percent_null', ascending=False)
)

application_null
```

```python
application_null.head(60)
```

Around 50 of the missing values are about information about building where the client lives. Probably can park it for the analysis?

```python
bureau_null = (
    bureau.isnull().sum()
          .to_frame(name='sum_of_nulls')
          .assign(percent_null=lambda d: d['sum_of_nulls'] * 100 / len(bureau))
          .reset_index(names='column')
          .sort_values('percent_null', ascending=False)
)
bureau_null
```

```python
previous_application_null_values = (
    previous_application.isnull().sum()
        .to_frame(name='sum_of_nulls')
        .assign(percent_null=lambda d: d['sum_of_nulls'] * 100 / len(previous_application))
        .reset_index(names='column')
        .sort_values('percent_null', ascending=False)
)
previous_application_null_values
```

It looks like a lot of features are null values, let's slowly figure out if the fields are necessary.
